```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#import txt file 
filename = 'eta_fz_values.txt'

with open(filename,'r') as f:
    lines=f.readlines()
print(lines)

eta=np.array(lines,dtype=float)

```

```python
#import txt file
filename = 'z_values.txt'

with open(filename,'r') as f:
    lines=f.readlines()
print(lines)

z=np.array(lines,dtype=float)
```

```python
plt.plot(z,eta)
plt.xlabel('z')
plt.ylabel('eta')
plt.title('eta vs z')

plt.show()
```

```python
#import txt file
filename = 'alpha_values.txt'

with open(filename,'r') as f:
    lines=f.readlines()
print(lines)

alpha=np.array(lines,dtype=float)
```

```python
plt.plot(z,alpha)
plt.xlabel('z')
plt.ylabel('alpha')
plt.title('alpha vs z')
plt.show()
```
