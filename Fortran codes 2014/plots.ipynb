```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#import txt file 
filename = 'eta_fz_values.txt'

with open(filename,'r') as f:
    lines=f.readlines()
print(lines)

eta=np.array(lines,dtype=float)

```

```python
#import txt file
filename = 'z_values.txt'

with open(filename,'r') as f:
    lines=f.readlines()
print(lines)

z=np.array(lines,dtype=float)
```

```python
plt.plot(z,eta)
plt.xlabel('z')
plt.ylabel('eta')
plt.title('eta vs z')
plt.xlim(-15,15)
plt.show()
```

```python
#import txt file
filename = 'alpha_values.txt'

with open(filename,'r') as f:
    lines=f.readlines()
print(lines)

alpha=np.array(lines,dtype=float)
```

```python
plt.plot(z,alpha)
plt.xlabel('z')
plt.ylabel('alpha')
plt.title('alpha vs z')
plt.xlim(-15,15)
plt.show()
```

```python
#import txt file
filename1 = 'Br_ini.txt'
filename2 = 'B_phi_ini.txt'



with open(filename1,'r') as f:
    lines1=f.readlines()
print(lines1)
with open(filename2,'r') as f:
    lines2=f.readlines()
print(lines2)
Br=np.array(lines1,dtype=float)
Bphi=np.array(lines2,dtype=float)
```

```python
plt.plot(z,Br)
plt.plot(z,Bphi)
plt.xlabel('z')
plt.ylabel('Br, Bphi')
plt.title('Br, Bphi vs z')
plt.show()
```

```python
#import txt file
filename1 = 'Br_final.txt'
filename2 = 'B_phi_final.txt'
filename3 = 'time.txt'




with open(filename1,'r') as f:
    lines1=f.readlines()
print(lines1)
with open(filename2,'r') as f:
    lines2=f.readlines()
print(lines2)
with open(filename3,'r') as f:
    lines3=f.readlines()
print(lines3)

Br_list=np.array(lines1,dtype=float)
Bphi_list=np.array(lines2,dtype=float)
time_list=np.array(lines3,dtype=float)
```

```python
plt.plot(z, Br_list, label='Br')
plt.plot(z, Bphi_list, label='Bphi')
plt.xlabel('z')
plt.ylabel('Br, Bphi')
plt.title(f'Br, Bphi vs z at t={time_list[-1]}')
plt.legend()
plt.show()
```
